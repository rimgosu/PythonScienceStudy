#### 특성확장

```python
col = X.columns
col
```

```python
for i in range(col.size): # 전체 컬럼
    for j in range(i, col.size): # 곱할 컬럼
        # CRIM*CRIM
        X[col[i]+'*'+col[j]] = X[col[i]] * X[col[j]] # 컬럼끼리 곱셈
col.size, X.shape
# 13 컬럼 => 104 컬럼으로 확장
```

- 선형모델(회귀) 평가 지표<br>
    : 여러 머신러닝 모델들을 사용했을 때 성능지표를 보고 모델을 비교하여 판단<br>
    : 정량적인 지표가 중요<br>
    : MSE : 0\~무한대,<br>
            ex) 오차 300 ==> 오차를 판단하는 기준 모호<br>
      MSE를 기준으로 R2 score를 사용 ==> score<br>
      R2 score : 분산을 기반으로 예측 성능 평가 (-1~1)<br>
      - -1 : 모델이 설명을 잘 못하는 경우<br>
      - 1 : 모델이 잘 설명함, 모델이 좋다<br>

### 규제를 적용시킨 모델
- LinearRegression 모델이 가지고 있는 단점
    - 모델이 잘못 되었을 경우 수정이 안된다 ==> 규제를 가해서 단점을 해소
    - LinearRegression + 규제(L1, L2)
    - L1규제 ==> Lasso
    - L2규제 ==> Ridge
    
- 규제? 선형모델에서 가중치(w)에 영향을 주는 것 ==> 모델에 개입을 한다.

```python
from sklearn.linear_model import Lasso, Ridge
```

#### Ridge 활용하기

```python
# 하이퍼파라미터(alpha) 튜닝
# 기본적으로 alpha값 = 1.0

def ridge_alpha(alpha):
    # ridge 모델 생성
    ridge = Ridge(alpha=alpha) # alpha값 범위 : -무한대 ~ 무한대
                               # 일반적으로 0.001 ~ 1000
    # 학습
    ridge.fit(X_train, y_train)
    # 결과 확인 - 훈련용데이터, 검증용데이터
    print(f'alpha : {alpha}', 'train_score : ', ridge.score(X_train, y_train))
    print('test_score : ', ridge.score(X_test, y_test))
    
```

```python
ridge_alpha(0.001)
ridge_alpha(0.01)
ridge_alpha(0.1)
ridge_alpha(1.0)
ridge_alpha(10)
ridge_alpha(100)
ridge_alpha(1000)
# 0.001일때 test_score 최대
```

```python
ridge_alpha(0.1)
# alpha 값을 증가 ==> 규제를 늘리겠다 : 모델이 복잡해지는 것을 막겠다.(과대적합일때)
# alpha 값을 감소 ==> 규제를 줄이겠다 : 모델이 단순해지는 것을 막겠다.(과소적합일때)
```

- Ridge (LinearRegression + L2규제) 정리
    - 규제를 줄이니까(alpha값 감소) test_score 상승
    - 과대적합보다는 과소적합 or 일반화에 가까운 상황
    - 모든 w에 특정 % 만큼의 규제를 가함 ==> 0에 가깝게는 되지만 0은 되지 않음
    - 규제를 많이 가하더라도 전체 가중치를 사용
    - 전체 데이터가 고루 중요할 때 사용

- Lasso 활용하기

```python
import numpy as np
```

```python
def lasso_alpha(alpha):
    lasso = Lasso(alpha=alpha) # alpha값 범위 : -무한대 ~ 무한대
                               # 일반적으로 0.001 ~ 1000
    # 학습
    lasso.fit(X_train, y_train)
    # 결과 확인 - 훈련용데이터, 검증용데이터
    print(f'alpha : {alpha}')
    print('train_score : ', lasso.score(X_train, y_train))
    print('test_score : ', lasso.score(X_test, y_test))
    
    # 사용한 특성의 수 확인
    print('사용한 특성의 수 : ' , np.sum(lasso.coef_ !=0))
    print()
```

```python
# np.sum(lasso.coef_ != 0)

# np.sum(값) ==> 값을 전부 다 더해주겠다.ㅈ
# lasso.coef_ ==> lasso 학습 후 전체 데이터 특성들의 가중치값을 출력
# 사용되지 않는 값(가중치) = 0 ==> False
# 사용된 값(가중치) !=0 ==> False
np.sum([True, False, True, False])
```

```python
lasso_alpha(0.001)
lasso_alpha(0.01)
lasso_alpha(0.1)
lasso_alpha(1.0)
lasso_alpha(10)
lasso_alpha(100)
lasso_alpha(1000)
# 0.01일때 test_score 최대
# 평균적으로 Ridge모델보다 성능 살짝 떨어진다.
```

- Lasso 모델 (L1규제)
    - 규제(증가)를 적용할수록 test score 값이 낮아짐
    - 규제를 적용할수록 사용하는 특성의 수도 감소
    
    - 모든 w에 특정 값만큼의 규제를 가함(상수)
    - 가중치가 0이되어 사용하지 않는 특성이 생김 ==> 특성 선택에 활용
    - 가중치가 -가 되면 0으로 처리 (사용하지 않는 특성으로 처리하겠다)
    - "특정 데이터가 중요할 때 사용"
