```python
from sklearn.linear_model import Lasso, Ridge
```

#### Ridge 

```python
# (alpha) 
#  alpha = 1.0

def ridge_alpha(alpha):
    # ridge  
    ridge = Ridge(alpha=alpha) # alpha  : - ~ 
                               #  0.001 ~ 1000
    # 
    ridge.fit(X_train, y_train)
    #   - , 
    print(f'alpha : {alpha}', 'train_score : ', ridge.score(X_train, y_train))
    print('test_score : ', ridge.score(X_test, y_test))
    
```

```python
ridge_alpha(0.001)
ridge_alpha(0.01)
ridge_alpha(0.1)
ridge_alpha(1.0)
ridge_alpha(10)
ridge_alpha(100)
ridge_alpha(1000)
# 0.001 test_score 
```

```python
ridge_alpha(0.1)
# alpha   ==>   :    .()
# alpha   ==>   :    .()
```

- Lasso 

```python
import numpy as np
```

```python
def lasso_alpha(alpha):
    lasso = Lasso(alpha=alpha) # alpha  : - ~ 
                               #  0.001 ~ 1000
    # 
    lasso.fit(X_train, y_train)
    #   - , 
    print(f'alpha : {alpha}')
    print('train_score : ', lasso.score(X_train, y_train))
    print('test_score : ', lasso.score(X_test, y_test))
    
    #    
    print('   : ' , np.sum(lasso.coef_ !=0))
    print()
```

```python
# np.sum(lasso.coef_ != 0)

# np.sum() ==>    .
# lasso.coef_ ==> lasso       
#   () = 0 ==> False
#  () !=0 ==> False
np.sum([True, False, True, False])
```

```python
lasso_alpha(0.001)
lasso_alpha(0.01)
lasso_alpha(0.1)
lasso_alpha(1.0)
lasso_alpha(10)
lasso_alpha(100)
lasso_alpha(1000)
# 0.01 test_score 
#  Ridge   .
```
